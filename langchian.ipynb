```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("GOOGLE_API_KEY")
if not api_key:
    raise ValueError("GOOGLE_API_KEY is not set. Add it to the .env file.")

llm_model = ChatGoogleGenerativeAI(
    google_api_key=api_key,
    temperature=0.6,
    model="gemini-2.5-flash",
)
text = "What is the difference between LangChain and OpenAI?"
response = llm_model.invoke(text)
print(response.content)
```

# Sequential Chain
When we use sequential Chain so if we use two or more templates we use sequential chain

```python
from langchain.chain import SequentialChain

capital_chain =PromptTemplate(
    input_variables=['country'],
    template ="What is the capital of {country}?"
)
capital_prompt = capital_chain.format_prompt(country="France")

famous_chain =PromptTemplate(
    input_variables=['capital'],
    template="What is {capital} famous for?"
)
famous_prompt = famous_chain.format_prompt(capital="Paris")

chain_model= SequentialChain(
    chains=[capital_prompt, famous_prompt], # we can add more chains here 
    input_variables=['country'],
    output_variables=['capital', 'famous_for']
)
chain_model.invoke({"country": "France"})
```
